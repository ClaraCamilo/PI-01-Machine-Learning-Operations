Creamos la función para desanidar


```python
def desanidar(df, columna, id, name):
    # Crear una lista vacía para almacenar los datos
    new_data = []

    # Iterar sobre cada registro de la columna  del DataFrame original
    for row in df[columna]:
        # Verificar si el valor es una lista
        if isinstance(row, list):
            # Crear una lista de los id y nombres correspondientes al registro actual
            ID = [a[id] for a in row]
            Nombres = [a[name] for a in row]

            # Agregar los datos a la lista
            new_data.append({f'Id_{columna}': ID, f'name_{columna}': Nombres})

    # Crear el nuevo DataFrame
    df_desanidado = pd.DataFrame(new_data)
    
    # Devolver el nuevo DataFrame
    return df_desanidado
```

```python
#Creamos un diccionario de los generos con el id y el nombre del genero 

# Crear una lista vacía para almacenar los datos únicos
unique_genres = []

# Iterar sobre cada registro en la columna "genres" del DataFrame original
for row in df2['genres']:
    # Iterar sobre cada género en el registro actual
    for genre in row:
        # Obtener el ID y nombre del género
        genre_id = genre['id']
        genre_name = genre['name']
        
        # Verificar si el género ya está en la lista de géneros únicos
        if genre_id not in [g['id'] for g in unique_genres]:
            # Agregar el género a la lista de géneros únicos
            unique_genres.append({'id': genre_id, 'name': genre_name})
# Ordenar la lista de géneros únicos por ID de menor a mayor
unique_genres = sorted(unique_genres, key=lambda x: x['id'])

# Crear el nuevo DataFrame a partir de la lista de géneros únicos
diccionario_generos = pd.DataFrame(unique_genres)
```

Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, dividiendo estas dos últimas revenue / budget, cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0.

```python
#Miramos el type de ambas columnas para verificar si necesitan ser modificados
df2["revenue"].info()
```

```python
df2["budget"].info()
```

```python
#Convertimos los valores no válidos o Nan a cero para poder hacer las conversiones necesarias
df2["budget"] = pd.to_numeric(df2["budget"], errors='coerce').fillna(0)
df2["revenue"].fillna(0, inplace= True)
```

```python
#Cambiamos el tipo de dato de budget para poder crear la nueva columna
df2["budget"] = df2["budget"].astype(float)
```

```python
# Creamos la columna con las especificaciones pedidas
df2["return"] = ''  # Crear una columna vacía

for fila in df2.index:
    #Si el valor de una de las dos columnas es 0, el resultado total será cero 
    if (df2.loc[fila, "revenue"] == 0) or (df2.loc[fila, "budget"] == 0):
        df2.loc[fila, "return"] = 0
    else:
        df2.loc[fila, "return"] = df2.loc[fila, "revenue"] / df2.loc[fila, "budget"]
```
